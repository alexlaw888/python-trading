```python
import numpy as np
import pandas as pd
from numpy.random import randn

#pandas must be imported along with numpy
```

```python
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df

#A-E are the row and W-Z are the columns(series)
```

```python
#Check for boolean condition

df > 0
```

```python
#Pass the boolen condition into booldf

booldf = df > 0
booldf
```

```python
#Pass back the booldf into the dataframe

df[booldf]
```

```python
#To simplify, simply insert the condition within the dataframe

df[df>0]
```

```python
#Determine if row is negative (true or false)

df['W']>0
```

```python
#Only return rows that are True(non negative)

df[df['W']>0]
```

```python
#Get all rows where Z is less than zero

df[df['Z']<0]
```

```python
#Row C excluded

resultdf = df[df['W']>0]
resultdf
```

```python
#With row C already excluded from the end result

resultdf['X']
```

```python
#Same as above but shorter form (a compsite result in one step)

df[df['W']>0]['X']
```

```python
#Can also get multiple list

df[df['W']>0][['Y','X']]
```

```python
#Identify that row C is negative

boolser = df['W']>0
boolser
```

```python
#Exclude row C and only take columns Y and X

boolser = df['W']>0
result = df[boolser]
result[['Y','X']]
```

```python
#Same as above breaking down into multiple steps

boolser = df['W']>0
result = df[boolser]
mycols = ['Y','X']
result[mycols]
```

```python
#Two or more conditions
# Note that df[(df['W']>0) and df['Y']>1] will generate an error, need to use &

df[(df['W']>0) & df['Y']>1]
```

```python
#Use pipe | instead of OR operator

df[(df['W']>0) | df['Y']>1]
```
